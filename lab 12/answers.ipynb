```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

pos_points = [[4, 1], [4, -1], [6, 0]]
neg_points = [[1, 0], [0, 1], [0, -1]]

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
plt.scatter([x[0] for x in pos_points], [x[1] for x in pos_points], color = 'red')
plt.scatter([x[0] for x in neg_points], [x[1] for x in neg_points], color = 'blue')
plt.show()
```

```python
# analyse the above plot to find the support vectors
pos_support_vec = [[4, -1], [4, 1]]
neg_support_vec = [[1, 0]]
vectors = pos_support_vec + neg_support_vec
homogenous_vec = np.array([x + [1] for x in vectors])

n = len(vectors)

A = homogenous_vec @ np.transpose(homogenous_vec)
B = np.array([1] * len(pos_support_vec) + [-1] * len(neg_support_vec))
alpha = np.linalg.solve(A, B)

W = np.zeros((1, len(homogenous_vec[0])))
ind = 0
for x in homogenous_vec:
    W = W + (alpha[ind] * x)
    ind += 1
    
# the last index of W has bias
b0, b1 = W[0][0] / W[0][1], W[0][2] / W[0][1]
print(b0, b1)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
plt.scatter([x[0] for x in pos_points], [x[1] for x in pos_points], color = 'red')
plt.scatter([x[0] for x in neg_points], [x[1] for x in neg_points], color = 'blue')
plt.plot([i for i in range(-5, 6)], [(b0 * x) + b1 for x in range(-5, 6)], color = 'yellow')
plt.show()
```
